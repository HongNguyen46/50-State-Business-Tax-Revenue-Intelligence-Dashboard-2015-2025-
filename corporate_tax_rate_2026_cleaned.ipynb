```python
import pandas as pd
import os
import glob
import re
```

```python
import glob

files = glob.glob("data_raw/*")
print(files)
```

```python
# Read the Excel file
df = pd.read_excel(
    r"data_raw\2025 State Corporate Income Tax Rates Brackets.xlsx",
    header=1
)

print(df.columns)
print(df.head())
```

```python
# Clean State names
# Remove footnotes such as (a), (b), (c)
# -----------------------------
df["State"] = (
    df["State"]
    .str.replace(r"\s*\(.*?\)", "", regex=True)
    .str.strip()
)
```

```python
print(df.columns)
```

```python
print(df[["State", "Rates"]].head(20))
```

```python
# Highest rate per state
highest_rate = (
    df.groupby("State", as_index=False)["Rates"]
      .max()
)
```

```python
# Fill missing rates with 0
df["Rates"] = df["Rates"].fillna(0)
```

```python
print(highest_rate.head(20))
```

```python
#Save the cleaned file
highest_rate.to_csv(
    "data_cleaned/corporate_tax_rate_2026_cleaned.csv",
    index=False
)
```
