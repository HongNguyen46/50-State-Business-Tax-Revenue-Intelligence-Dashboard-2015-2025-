**DEVELOP AN AUTOMATED PROCESS TO CLEAN THE REMAINING 26 EXCEL FILES**

```python
import os
import re
import glob
import pandas as pd

# Create output folder
os.makedirs("data_cleaned", exist_ok=True)

# Get all QTAX files
files = sorted(glob.glob("data_converted/qtax_*.xlsx"))

for file in files:

    print(f"Processing {os.path.basename(file)}...")

    # -----------------------------
    # Extract year and quarter
    # -----------------------------
    match = re.search(r"qtax_(\d{4})_(q\d)\.xlsx", os.path.basename(file), re.IGNORECASE)

    year = int(match.group(1))
    quarter = match.group(2).upper()

    # -----------------------------
    # Read Excel
    # -----------------------------
    df = pd.read_excel(file, header=5)

    # Remove unnamed columns
    df = df.loc[:, ~df.columns.str.contains("^Unnamed")]

    # Convert values to numeric
    value_cols = df.columns.drop(["Tax Description", "Code"])
    df[value_cols] = df[value_cols].apply(pd.to_numeric, errors="coerce")

    # -----------------------------
    # Reshape to long format
    # -----------------------------
    state_cols = [
        c for c in value_cols
        if "U.S. Total" not in c
    ]

    df_long = df.melt(
        id_vars=["Tax Description", "Code"],
        value_vars=state_cols,
        var_name="State",
        value_name="Tax_Value"
    )

    # Clean tax description
    df_long["Tax Description"] = (
        df_long["Tax Description"]
        .str.replace("\xa0", "", regex=True)
        .str.strip()
    )

    # -----------------------------
    # Keep only desired taxes
    # -----------------------------
    keep_taxes = [
        "Property taxes",
        "General sales and gross receipts",
        "Individual income",
        "Individual income taxes",
        "Corporation net income",
        "Corporate income taxes"
    ]

    df_long_clean = df_long[
        df_long["Tax Description"].isin(keep_taxes)
    ].copy()

    # Standardize tax names
    df_long_clean["Tax_Group"] = df_long_clean["Tax Description"].replace({
        "Property taxes": "Property Tax",
        "General sales and gross receipts": "Sales Tax",
        "Individual income": "Individual Income Tax",
        "Individual income taxes": "Individual Income Tax",
        "Corporation net income": "Corporate Income Tax",
        "Corporate income taxes": "Corporate Income Tax"
    })

    # -----------------------------
    # Pivot to final format
    # -----------------------------
    df_final = (
        df_long_clean
        .pivot_table(
            index="State",
            columns="Tax_Group",
            values="Tax_Value",
            aggfunc="sum"
        )
        .reset_index()
    )

    # Ensure all expected columns exist
    expected_cols = [
        "Property Tax",
        "Sales Tax",
        "Individual Income Tax",
        "Corporate Income Tax"
    ]

    for col in expected_cols:
        if col not in df_final.columns:
            df_final[col] = 0

    # Add year and quarter
    df_final["Year"] = year
    df_final["Quarter"] = quarter

    # Total tax
    df_final["Total Tax"] = (
        df_final["Property Tax"] +
        df_final["Sales Tax"] +
        df_final["Individual Income Tax"] +
        df_final["Corporate Income Tax"]
    )

    # Arrange columns
    df_final = df_final[
        [
            "State",
            "Year",
            "Quarter",
            "Property Tax",
            "Sales Tax",
            "Individual Income Tax",
            "Corporate Income Tax",
            "Total Tax"
        ]
    ]

    # -----------------------------
    # Save cleaned file
    # -----------------------------
    output_file = os.path.join(
    "data_cleaned",
    os.path.basename(file).replace(".xlsx", "_clean.csv")
)

    df_final.to_csv(output_file, index=False)

print("\nAll files have been processed successfully!")
```
