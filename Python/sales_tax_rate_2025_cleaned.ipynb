```python
# Read the Excel file
df = pd.read_excel(
    r"data_raw\2025 Sales Tax Rates as of July 1, 2025.xlsx",
    header=1
)

print(df.columns)
print(df.head())
```

```python
# Clean State names
# Remove footnotes such as (a), (b), (c)
# -----------------------------
df["State"] = (
    df["State"]
    .str.replace(r"\s*\(.*?\)", "", regex=True)
    .str.strip()
)

# Remove all non-state rows (notes/footer)
df = df[
    ~df["State"].str.contains(
        r"Three states levy|The sales taxes|Special taxes|Salem County|Sources:|Note:",
        case=False,
        na=False
    )
]

df = df[df["State"].notna()]
df = df[df["State"].str.len() > 0]
```
